```
#import nltk library
import nltk
```

***Downloading all packages from NLTK***

```
nltk.download()
```

***Fetch a URL and print***

```
import urllib.request
response =  urllib.request.urlopen('https://en.wikipedia.org/wiki/Natural_Language_Toolkit')
html = response.read()
#print(html)
```

**BeautifulSoup library to pull data from HTML files**

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip = True)
print(text)
```

**Performing tokenization and printing the tokens**

```
tokens = [t for t in text.split()]
print(tokens)
```

**Pre-processing by removing multiple tokens and plot Frequency Distribution**

```
from nltk.corpus import stopwords
sr= stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
      clean_tokens.remove(token)
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))
freq.plot(20, cumulative=False)
```
